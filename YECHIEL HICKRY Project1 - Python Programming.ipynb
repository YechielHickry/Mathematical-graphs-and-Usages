## Part I – The Node class
### **Task 1 – Define the class**  
Implement the Node class

```python
from datetime import datetime
from os import getcwd

# ************************************************************
# *********************  part I   ****************************
# ************************************************************
print "{} \n part I \n{} \n".format('*'*50, '*'*50)


class Node:
    def __init__(self, node_name):
        """
        The method does not have to test
        the validity of name.
        :param: name
        """
        self.name = node_name
        self.neighbors = {}

    def __str__(self):
        """
        representation og the node
        :return: string graph name + neighbors
        """
        return "{:3} neighbors -> {} ".format(str(self.name),
                                              str([(neighbor, self.neighbors[neighbor])
                                                   for neighbor in self.neighbors]))

    def __len__(self):
        """
        :return: the number of neighbors
        """
        return len(self.neighbors)

    def __contains__(self, item):
        """
        :type item: string
        :param item:
        :return: Boolean ,True if item is a name of a neighbor of self
        """
        return True if item in self.neighbors else False

    def __getitem__(self, key):
        """
        :param key:
        :return: returns the weight of the neighbor named key.
        If there is no such neighbor, then the method returns None
        """
        return self.neighbors[key] if key in self.neighbors else None

    def __eq__(self, other):
        """
        :param other:
        :return: boolean, True if  self.name == other.name
        """
        return self.name == other.name

    def __ne__(self, other):
        """
        :param other:
        :return: boolean, True if  self.name != other.name
        """
        return not (self == other)

    def is_neighbor(self, name):
        """
        :param name:
        :return:  boolean, True if name is a neighbor of self
        """
        return name in self

    def add_neighbor(self, neighbor_name, weight=1):
        # type: (object, object) -> object
        """
        This method does not test whether a node named name exists.
        This method should not allow adding a neighbor with the same name as self.
        using __contains__(self,item):
        :param neighbor_name
        :param weight
        :return:boolean, true if name has been added to self, Else False
        """
        if neighbor_name == self.name:
            print "a node can not be his own neighbor"
            return False
        else:
            self.neighbors[neighbor_name] = weight
            return True

    def remove_neighbor(self, name):
        """
        removes name from being a neighbor of self.
        This method does not test whether a node named name exists
        This method should not fail if name is not a neighbor of self
        :param name:
        :return: weight if name is neighbor of self else None
        """
        return self.neighbors.pop(name, None)

    def get_weight(self, name):
        """
        using __getitem__(self, key)
        :param name:
        :return: returns the weight of the neighbor named key.
        If there is no such neighbor, then the method returns None
        """
        return self[name]

    def is_isolated(self):
        """
        :return: True if self has no neighbors
        """
        return len(self) == 0

    def total_edge_weight(self):
        """
        :return: tuple (self.number of edges,self.sum of weights)
        """
        num_edge = 0
        total_weight = 0
        for weight in self.neighbors.values():
            num_edge += 1
            total_weight += weight
        return num_edge, total_weight
```

## Part II – The Graph class
### **Task 1 – Define the class**  
Implement the Graph class with the following properties:

```python
# ************************************************************
# *********************  part II  ****************************
# ************************************************************
print "{} \n part II \n{} \n".format('*' * 50, '*' * 50)


class Graph:
    def __init__(self, name, nodes=None):
        """
        nodes is an iterable of Node instances.
        Note: nodes is an iterable (preferably list) of Node instances,
        while the attribute nodes is a dictionary. This is not a mistake.
        """
        if nodes is None:
            nodes = []
        self.name = name
        self.nodes = dict(map(lambda x: (x.name, x), nodes))

    def __str__(self):
        """
        :return: string -description of all the nodes in the graph
        """
        my_string = "Graph name : {} \n".format(self.name)
        my_string += '\n'.join([str(node) for node in self.nodes.values()])
        return my_string

    def __len__(self):
        """
        :return: the number of nodes in the graph
        """
        return len(self.nodes)

    def __contains__(self, key):
        """
        returns True in two cases:
         (1) If key is a string, then if a node called key is in self, and
         (2) If key is a Node, then if a node with the same name is in self.
        :param key:
        :return: boolean
        """
        if isinstance(key, Node):
            return key in self.nodes.values()
        else:
            return key in self.nodes

    def __getitem__(self, name):
        """
        This method should raise KeyError if name is not in the graph
        :param name:
        :return: the Node object whose name is name
        """
        try:
            return self.nodes[name]
        except KeyError:
            raise "KeyError '{}' - is not a node in graph '{}'  ".format(name, self.name)

    def add_node(self, node):
        """
        adds a new node to the graph
        If a node with the same name already exists in self,existing edges should be overwritten,
        and new edges should be created
        If node has neighbors that are not already in self, then this method should create the relevant nodes
        :param node: Node instance
        :return:
        """
        if node.name not in self:
            self.nodes[node.name] = node
        else:
            for neighbor_name in node.neighbors:
                self[node.name].add_neighbor(neighbor_name, node[neighbor_name])

        # If node has neighbors that are not already in self, then this method should create the relevant nodes
        for neighbor in node.neighbors:
            if neighbor not in self:
                self.nodes[neighbor] = Node(neighbor)

    def __add__(self, other):
        """
        :param other:
        :return: a new Graph object that includes all the nodes and edges of self and other
        """
        new_graph = Graph(self.name + ' + ' + other.name, [])
        for node in self.nodes.values():
            new_graph.add_node(node)
        for node in other.nodes.values():
            new_graph.add_node(node)
        return new_graph

    def remove_node(self, name):
        """
        This method should not fail if name is not in self
        This method should also remove edges, in which name is a neighbor of other nodes in the graph.
        :param name:
        :return: weight if name is neighbor of self else None

        """
        res = self.nodes.pop(name, None)
        for my_node in self.nodes.values():
            my_node.remove_neighbor(name)
        return res

    def is_edge(self, frm_name, to_name):
        """
        :param frm_name:
        :param to_name: the neighbor of frm_name
        :return:True if to_name is a neighbor of frm_name else False
        """
        return self[frm_name].is_neighbor(to_name) if (frm_name in self) and (to_name in self) else False

    def add_edge(self, frm_name, to_name, weight=1):
        # type: (object, object, object) -> object
        """
        adds an edge making to_name a neighbor of frm_name.
        This method should not fail if either frm_name or to_name are not in self.
        :param frm_name:
        :param to_name:
        :param weight: default = 1
        :return:
        """
        if frm_name not in self:
            self.nodes[frm_name] = Node(frm_name)
            self.add_edge(frm_name, to_name, weight)  # recursive
            return
        else:
            self.nodes[frm_name].add_neighbor(to_name, weight)
        if to_name not in self:
            self.nodes[to_name] = Node(to_name)

    def remove_edge(self, frm_name, to_name):
        """
        removes to_name from being a neighbor of frm_name.
        This method should not fail if frm_name is not in self.
        This method should not fail if to_name is not a neighbor of frm_name.
        :param frm_name:
        :param to_name:
        :return: weight of edge in case the edge exist else none
        """
        return self.nodes[frm_name].remove_neighbor(to_name) if self.is_edge(frm_name, to_name)else None

    def get_edge_weight(self, frm_name, to_name):
        # type: (object, object) -> object
        # type: (object, object) -> object
        """
        This method should not fail if either frm_name or to_name are not in self
        This method should return None if to_name is not a neighbor of frm_name.
        :param frm_name:
        :param to_name:
        :return: weight of the edge between frm_name and to_name, /
        None if frm_name or to_name are not in self./
        This method should return None if to_name is not a neighbor of frm_name.
        """
        return self.nodes[frm_name].get_weight(to_name) if self.is_edge(frm_name, to_name)else None

    def get_path_weight(self, path):
        """
        Tip: The built-in functions any() and all() regard nonzero numbers as True and None as False.
        :param path: list,is an iterable of nodes names.
        :return: the total weight of the given path /
        This method should return None if the path is not feasible in self /
        This method should return None if path is an empty iterable.
        """
        lst_edges = zip(path[:-1], path[1:])
        path_weight = map(lambda (x, y): self.get_edge_weight(x, y), lst_edges)
        if all(path_weight):  # the path exist
            return reduce(lambda x, y: x + y, path_weight)
        else:
            return None

    def is_reachable(self, frm_name, to_name):
        """
        :param frm_name:
        :param to_name:
        :return:
        """
        # Use BFS to check path between s and d
        # Mark all the vertices as not visited
        nodes_not_visited = list(self.nodes)

        # Create a queue for BFS
        queue = [frm_name]
        # Mark the source node as visited and enqueue it
        # queue.append(frm_name)
        nodes_not_visited.remove(frm_name)
        while queue:

            # pop a vertex from queue
            n = queue.pop(0)

            # If this adjacent node is the destination node,
            # then return true
            if n == to_name:
                return True

            # Else, continue to do BFS
            for neighbor_name in self.nodes[n].neighbors:
                if neighbor_name in nodes_not_visited:
                    queue.append(neighbor_name)
                    nodes_not_visited.remove(neighbor_name)
                    # If BFS is complete without visited d
        return False

    def find_shortest_path_all(self, frm_name):
        """
        :param frm_name:
        :return:distances is a dict that contains the min()
        weight from frm_name to all the vertex' if ==-1 the not is_reachable
                 path_frm_name a dict that contains shortest path from frm_name
        """
        # Use BFS to check path between s and d
        # Mark all the vertices as not visited
        nodes_not_visited = list(self.nodes)

        distances = dict(zip(list(self.nodes), [-1] * len(self)))
        path_frm_name = dict(zip(list(self.nodes), [[]] * len(self)))

        # Create a queue for BFS
        queue = [frm_name]

        # Mark the source node as visited and enqueue it
        nodes_not_visited.remove(frm_name)
        distances[frm_name] = 0
        path_frm_name[frm_name] = [frm_name]
        while queue:

            # pop a vertex from queue
            node_from_queue = queue.pop(0)

            # Else, continue to do BFS
            for neighbor_of_n in self.nodes[node_from_queue].neighbors:
                if neighbor_of_n in nodes_not_visited:  # once we checked the vertex there is no need to re-check it
                    queue.append(neighbor_of_n)
                    nodes_not_visited.remove(neighbor_of_n)
                frm_to_weight = int(self.get_edge_weight(node_from_queue, neighbor_of_n))
                if (distances[neighbor_of_n] == -1) or \
                        (distances[neighbor_of_n] > distances[node_from_queue] + frm_to_weight):
                    distances[neighbor_of_n] = distances[node_from_queue] + frm_to_weight
                    path_frm_name[neighbor_of_n] = path_frm_name[node_from_queue] + []  
                    /# update the shorter path
                    path_frm_name[neighbor_of_n].append(neighbor_of_n)

        return distances, path_frm_name

    def find_shortest_path(self, frm_name, to_name):
        if self.is_reachable(frm_name, to_name):
            all_distances, all_path_frm_name = self.find_shortest_path_all(
                frm_name)  # we assume that a node is reachable to itself
            return all_path_frm_name[to_name]

        else:
            return None
```

### Task 3 – The roadmap implementation  
The files travelsEW.csv and travelsWE.csv record a large number of

#### Question 1
From each file create a graph whose nodes are the country regions,and whose edges are the roads themselves (if a travel was not recorded between country regions, then it means such road does not exist). The weight of each edge is defined as the average time (in seconds) of all the travels done on that road. When the two graphs are ready,  
add them together to create the complete graph of the roadmap.

```python
# *************************************************************
# ************** The road-map implementation ******************
# *************************************************************

# The files travelsEW.csv and travelsWE.csv record a large number of travels made by people
# from five regions in the country, called Center, North, South, East and West.
f_name_ew = getcwd() + '\\travelsEW.csv'
f_name_we = getcwd() + '\\travelsWE.csv'
ew_date_format = "%d/%m/%Y %Hh%Mm"  # 01/01/2016 01h08m
we_date_format = "%I:%M:%S%p ; %b %d %y"  # 05:58:00PM ; Jan 07 16


def extract_road_trip_time(f_name, my_date_format):
    # type: (object, object) -> object
    # type: (object, object) -> object
    travel_list = []
    with open(f_name, 'r') as f:
        for i, line in enumerate(f, 1):  # Center,07/01/2016 01h13m,West,07/01/2016 02h34m
            # print i
            my_line = line.strip().split(',')  # [Center,07/01/2016 01h13m,West,07/01/2016 02h34m]
            if (i == 1) or (len(my_line) != 4):
                continue
            else:
                try:
                    start_t = datetime.strptime(my_line[1], my_date_format)
                    end_t = datetime.strptime(my_line[3], my_date_format)
                    travel_time = end_t - start_t
                except ValueError:  # Center,22/01/2016 19h30m,North,22/01/2016 25h52m
                    print ValueError
                except IndexError:  # Center,07/03/2016 03h14m,W
                    print IndexError
                travel_list.append(((my_line[0], my_line[2]), travel_time.total_seconds()))
    return list(travel_list)


def road_average_travel_time(road, data_list):
    """
    :param road: contains frm_city: origin, to_city:  destination
    :param data_list:  is a list of tuples containing all roads in the file
    :return:  int - this function return the average travel time between frm_city to to_city, ignoring all other roads
    """
    frm_city, to_city = road
    travel_time_list = filter(lambda x: x[0][0] == frm_city and x[0][1] == to_city, data_list)
    if len(travel_time_list) > 1:
        sum_time = sum(map(lambda x: x[1], travel_time_list))
        return int(round((sum_time / len(travel_time_list))))
    else:
        return None
# extracting the data from file , return (frm_city,to_city,travel_time_in sec
ew_travel_list = extract_road_trip_time(f_name_ew, ew_date_format)
roads_ew = set(map(lambda x: x[0], ew_travel_list))

# creating graph EW and then adding the edges
graph_ew = Graph('graph_ew', [Node(item) for item in ['Center', 'North', 'South', 'East', 'West']])
for elm in roads_ew:
    my_weight = road_average_travel_time(elm, ew_travel_list)
    if my_weight != None :
        graph_ew.add_edge(elm[0],elm[1],my_weight)
# print graph_ew

# extracting the data from file , return (frm_city,to_city,travel_time_in sec
we_travel_list = extract_road_trip_time(f_name_we, we_date_format)
roads_we = set(map(lambda x: x[0], we_travel_list))
# creating graph WE and then adding the edges
graph_we = Graph('graph_we', [Node(item) for item in ['Center', 'North', 'South', 'East', 'West']])
for elm in roads_we:
    my_weight = road_average_travel_time(elm, we_travel_list)
    if my_weight != None :
        graph_we.add_edge(elm[0],elm[1],my_weight)
# print graph_we

# creating graph NEW which include both EW & WE graphs
graph_new = graph_ew + graph_we
print "road map question 1 :\n{}".format(graph_new)
```

```python
# ************************************************************
# *********************  part III  ***************************
# ************************************************************
print "{} \n part III \n{} \n".format('*' * 50, '*' * 50)


class NonDirectionalGraph(Graph):
    def __init__(self, name, nodes=None):
        """
        nodes is an iterable of Node instances.
        Note: nodes is an iterable (preferably list) of Node instances,
        while the attribute nodes is a dictionary. This is not a mistake.
        keep in mind ! that if the graph get a directional nodes it changes them to non directional nodes
        """
        self.name = name
        self.nodes = {}
        for node in nodes:
            self.add_node(node)
        # self.directional = False

    def add_node(self, node):
        """
        adds a new node to the graph
        graph.add_node deals with :
            If a node with the same name already exists in self,existing edges should be overwritten,
            and new edges should be created
            If node has neighbors that are not already in self, then this method should create the relevant nodes
        in addition, updating or adding additional edge from neighbor to node
        node-> {neighbor,weight} => neighbor-> {node,weight}
        :param node: Node instance
        :return:
        """
        Graph.add_node(self, node)
        # If node has neighbors that are not already in self, then this method should create the relevant nodes
        # updating or adding additional edge from neighbor to node
        for neighbor_name in node.neighbors:
            weight = node.neighbors[neighbor_name]
            to_name = node.name
            self.add_edge(neighbor_name, to_name, weight)

    def __add__(self, other):
        """
        :param other:
        :return: a new non directional graph object that includes all the nodes and edges of self and other
        """
        new_graph = NonDirectionalGraph(self.name + ' + ' + other.name,[])
        for node in self.nodes.values():
            new_graph.add_node(node)
        for node in other.nodes.values():
            new_graph.add_node(node)
        return new_graph

    def add_edge(self, frm_name, to_name, weight=1):
        """
        Graph.add_edge :
            adds an edge making to_name a neighbor of frm_name.
            if either frm_name or to_name are not in self the function create a Node and add it to graph
        in non  directional graph we should add the reverse edge as well
        :param frm_name:
        :param to_name:
        :param weight: default = 1
        :return:
        """
        Graph.add_edge(self, frm_name, to_name, weight)
        Graph.add_edge(self, to_name, frm_name, weight)

    def remove_edge(self, frm_name, to_name):
        """
        Graph.remove_edge :
            removes to_name from being a neighbor of frm_name.
            This method should not fail if frm_name is not in self.
            This method should not fail if to_name is not a neighbor of frm_name.
        in non  directional graph we should remove the reverse edge as well
        :param frm_name:
        :param to_name:
        :return: weight of edge in case the edge exist else none
        """
        r_edge_f_t = Graph.remove_edge(self,frm_name,to_name)
        r_edge_t_f = Graph.remove_edge(self,to_name,frm_name)
        return int((r_edge_f_t +r_edge_t_f)/2)  if r_edge_f_t and r_edge_t_f else None
```
